### Leitura de arquivos

````python
import csv

# Função que irá processar o arquivo
def read_csv(file_path):
    """Função que fará a leitura do arquivo e devolverá
    uma lista de dicionário com os dados do arquivo
    
    Argumentos:
    file_path -- Caminho do arquivo de entrada
    """
    
    out = []
    
    # Abertura do arquivo
    with open(file_path, encoding='utf-8-sig') as csvfile:
        reader = csv.DictReader(csvfile, delimiter=';')
        
        # Iteração nas linhas do arquivo
        for row in reader:
            out.append(row)
        
        return out
````

#### 1.4.2.3 - Dicionários - dict

Dicionários são estruturas em que os ínidices não são valores numéricos, e sim, qualquer valor textual. Dicionários são muito utilizado quando desejamos criar estrutras mais complexas e com mais informações. Por exmplo, digamos que agora, além do nome dos alunos, é necessário adicionarmos tambémas as 4 notas referentes as provas realizadas por esses alunos. Criar essa estrutura utilizando listas ou tuplas seria algo confuso e copmplicado de manter ao longo do tempo.

Dessa forma, podemos utilizar um mapa para criarmos a estrutura que representará o aluno:

```python
{'nome': 'Pedro', 'nota1': 10, 'nota2': 8, 'nota3': 7, 'nota4': 10}
```

E depois adicionar cada estrutura de aluno em uma lista, obtendo a seguinte estrutura:

````python
# Criando uma lista que conterá todos os alunos,
# sendo que cada aluno está representado utilizando
# a estrutura de dicionário
alunos = [
    {'nome': 'Pedro',     'nota1': 8, 'nota2': 8, 'nota3': 7, 'nota4': 9},
    {'nome': 'Guilherme', 'nota1': 9, 'nota2': 7, 'nota3': 6, 'nota4': 8},
    {'nome': 'Volpato',   'nota1': 1, 'nota2': 7, 'nota3': 7, 'nota4': 7},
    {'nome': 'Ferraresi', 'nota1': 2, 'nota2': 8, 'nota3': 6, 'nota4': 6},
]

alunos
````

Dicionários são estruturas compostas por `chave` e `valor`, onde a `chave` representa o índice ou o indicador do `valor`.

Os dicionários, assim como as listas, nos permitem alterar seus conteúdos. Para isso, basta acessarmos a `chave` que desejamos alterar e mudar o seu valor. Por exemplo, digamos que houve um erro de digitação na nota do aluno `Volpato`, e que o valor correto da chave `nota1` desse aluno seja o valor `10` ao ínves do valor `1`.

Para fazermos essa alteração, primeiro precisamos selecionar o aluno `Volpato` que está em nossa lista. Como sabemos que ele está no índice `2` da lista, basta acessarmos da seguinte maneira:

````python
aluno = alunos[2]
aluno
````

Agora que temos o aluno `Volpato` selecionado, vamos selecionar a chave `nota1` e alterar o seu valor:

````python
# Alteração do valor da chave 'nota1' do aluno
aluno['nota1'] = 10

# Exibindo as novas notas do aluno
aluno
````

Observe que a chave `nota1` foi alterada. 

Além disso, para inserirmos um novo valor em um dicionário, digamos a média das notas, basta adicionar o nome da `chave` que desejamos dentro dos colchetes, como se fossemos alterar essa `chave`. Por exemplo:

````python
# Inserindo a chave 'media' no dicionario
aluno['media'] = 7.75

# Exibindo o conteudo após a inserção da chave e seu valor
aluno
````

E para remover uma `chave` e seu respectivo valor, basta utilizarmos a função `pop()` passando como parâmetro o nome da `chave` que desejamos remover. Por exemplo:

````python
# Removendo a chave 'media' do dicionário
aluno.pop('media')

# Exibindo o conteúdo após a remoção da chave
aluno
````

Além disso,  outro ponto muito importante dos dicionários é que suas `chaves` são **únicas**. Ou seja, se dentro de um dicionário já houver uma `chave` chamada `media`, se tentarmos adicionar essa chave novamente, o que acontecerá é que o valor dessa `chave` já existente será alterado ou atualizado!

# 2.0 - Estrutura de Decisão - `if`

Agora que já revimos o básico do Pyhthon, como por exemplo como criar e inserir valores em variáies, e quais tipos de variáveis podemos criar, vamos revisar um dos principais pontos, senão o princiapal ponto do curso: Lógica e Resolução de Problemas!

E para iniciarmos essa revisão, temos que ser capazes de poder "questionar" ou "verificar" certas situações. Vamos pegar como exemplo o exercício anterior, onde tínhamos uma estrutura de dicionário contendo os dados de um aluno hipotético. Dentro dessa estrutura, tínhamos inicialmente quatro notas referentes a quatro testes distintos que esse aluno fez.

O exercício anterior pedia para que calculássemos a média e verificássemos o valor da média. Se a média fosse igual ou maior que 7, o aluno estaria aprovado, caso contrário, estaria reprovado.

No exercício anterior essa verificação da média foi feita de maneira *hard coded*, ou seja, nós mesmo verificamos o valor da média exibindo-a na tela, para então escrevermos se o aluno está ou não aprovado. Porém, existe uma maneira de fazermos esse processo "automaticamente", ou seja, o próprio programa irá decidir para nós se o aluno está ou não reprovado com base no valor da média.

Por exemplo, considere o seguinte aluno:

```python
{'nome': 'Pedro', 'nota_1': 10, 'nota_2': 9, 'nota_3': 8, 'nota_4': 10}
```

Se fossemos seguir o exemplo da aula anterior, teríamos o seguinte código para calcularmos a média das notas:

```python
aluno = {'nome': 'Pedro', 'nota_1': 10, 'nota_2': 9, 'nota_3': 8, 'nota_4': 10}

aluno['media'] = (aluno['nota_1'] + aluno['nota_2'] + aluno['nota_3'] + aluno['nota_4']) / 4

print(aluno['media']) # 9,25
```

Como fazer então para que a própria aplicação faça a verificação necessário da média e defina se o aluno está ou não aprovado? A resposta é simples: Utilizando a estrutura condicional `if`.

A estrutura `if` foi criada para testarmos e/ou verificarmos variáveis e/ou as relações entre as variáveis. Ou seja, através do `if`, podemos testar se o valor da média recém calculada é maior ou igual a 7. Caso seja, podemos criar a chave `aprovado` com o valor correto:

```python
if(aluno['media'] >= 7):
    aluno['aprovado'] = True
```

E caso o aluno não seja aprovado, podemos utilizar a cláusuila `else` que existe dentro do `if` para escrevermos que o aluno está reprovado. Dessa forma, temos o `if` que é uma cláusula que realiza um determinado teste ou verificação e, caso esse teste ou verificação seja verdadeira, executará algum comando; Temos também o `else`, que é a cláusula contrária ao `if`, ou seja, se o teste ou verificação não for verdadeira, o que será executado é o bloco da cláusula `else`.

Dessa forma, temos o seguinte código:

```python
if(aluno['media'] >= 7):
    aluno['aprovado'] = True
else
    aluno['aprovado'] = False
```

Agora imagine a seguinte situação: Ao invés de utilizarmos a chave `aprovado` que utiliza valores *booleanos*, como `True` e `False`, iremos criar uma nova chave, chamada `situacao`, que terá como conteúdo as palavras `aprovado`, caso o aluno seja aprovado; `recuperação`, caso o aluno esteja em recuperação e `reprovado` caso o aluno esteja reprovado. 

Para o aluno estar aprovado, a sua nota continua sendo sete ou mais. Para que ele esteja em recuperação, sua nota tem que estar entre 5 (inclusive) e menor que 7. E para que ele seja reprovado, a nota tem que ser menor que 5.

Como vimos anteriormente, com a cláusula `if` podemo ter uma ação tanto quando uma determinada situação é verdadeira quanto quando esse situação é falsa, mas somente testando uma única condição. E agora temos duas condições para testar: se o aluno possuí a média maior ou igual a sete, senão, temos que verificar se ele possuí a média maior que cinco.

E é nesse momento que utilizamos a cláusula `elif`. A cláusula `elif` nos permite criar quantos testes ou verificações precisarmos entre as cláusulas `if` e `else`. Dessa forma, nosso desafio seria solucionado da seguinte maneira:

```python
if(aluno['media'] >= 7):
    aluno['situacao'] = 'aprovado'
elif(aluno['media'] >=5):
    aluno['situacao'] = 'recuperação'
else:
    aluno['situacao'] = 'reprovado'
```

Dessa forma, a estrutura básico do `if` é:

```python
if (condicao_1):
    # Comandos caso a condição-1 seja verdadeira
elif (condicao_2):
    # Comandos caso a condição-2 seja verdadeira
else:
    # Comandos caso nenhuma das condições seja verdadeira
```

Agora que revisamos a estrutura do `if`, vamos revisar como montar ou compor as nossas condições. Uma condição, ou teste, nada mais é que uma verificação de algum estado, variável ou valor. E essas verificação são feitas utilizando operadores.

Existem dois tipos principais de operadores que precisamos rever para melhorar nossa capacidade de composição de condições, ou testes: 
- **Operadores Relacionais**
- **Operadores Lógicos**

## 2.1 - Operadores Relacionais

Operadores Relacionais são os operadores que nós utilizamos quando desejamos verificar a "relação" entre variáfeis, ou qual o estado de uma variável em relação à outra variável. Pense neles, nos operadores relacioanis, como formas de testarmos ou verificarmos a grandeza ou o estado de uma variável em relação a um determinado valor ou outra variável.

Por exemplo, quando verificamos se o a média das notas do aluno era maior ou igual a 7, utilizamos um operador relacional, o `maior ou igual (>=)`. Foi através desse operador que validamos que a média possuia um valor maior que 7, ou seja, verificamos o estado do valor que estava dentro da chave `media` em relação ao valor 7.

## 2.2 - Operadores Lógicos

Operadores Lógicos são os operadores que utilizamos para realizar testes lógicos ou [testes boleanos](https://libguides.mit.edu/c.php?g=175963&p=1158594). Esses operadores são muito utilizados quando desejamos criar uma condição que é um pouco mais complexa. Ou seja, que possua duas ou mais condições em conjunto.

Por exemplo, digamos que você possua o seguinte conjunto de dados:

```python
alunos = [
    {'nome': 'Pedro',     'nota1': 8, 'nota2': 8, 'nota3': 7, 'nota4': 9},
    {'nome': 'Guilherme', 'nota1': 9, 'nota2': 7, 'nota3': 9, 'nota4': 8},
    {'nome': 'Volpato',   'nota1': 1, 'nota2': 7, 'nota3': 7, 'nota4': 7},
    {'nome': 'Ferraresi', 'nota1': 2, 'nota2': 8, 'nota3': 6, 'nota4': 6},
]
```

Você deseja veriricar e separar o aluno que possuí as notas `nota_1` e `nota_3` iguais ou maiores que 9. Uma maneira de resolver esse desafio, sem a utilização de operadores lógicos seria encadear estruturas `if`, ou seja, colocar uma condição dentro da outra:

```python
if (aluno['nota_1'] >= 9):
    if(aluno['nota_3'] >= 9):
        # Comandos para separar o aluno
```

Porém, essa forma, além de mais trabalhosa, não é tão elegante. E é aí que utilizamos operadores lógicos. Quando desejamos testar a condição da `nota_1` e da `nota_3` em uma única cláusula `if`, facilitando assim a leitura e o entendimento. Dessa forma, a cláusula `if` ficaria da seguinte maneira:

```python
if (aluno['nota_1'] >= 9 and aluno['nota_3'] >= 9):
    # Comandos para separar o aluno
```

Observe que em uma única linha, estamos testando ambas as condições, e para isso, utilizamos o operador relacional `and`. Para melhorar o entendimento, imagine que a condição `aluno['nota_1'] >= 9` seja substituida por `x`, e que a condição `aluno['nota_3'] >= 9` seja substituida por `y`. Com isso, teríamos o seguinte resultado:

```python
if (x and y):
    # Comandos para separar o aluno
```

Observe na imagem abaixo o funcionamento do operador `and`:

````python
Image('img/op_and.png')
````

Com a **Tabela Verdade** acima, temos todos os resultados possíveis testando ambas as condições `x` e `y`. Observe que o resultado do teste só retorna verdadeiro (`True`) quando ambas as condições `x` e `y` são verdadeiras. E é por conta desse comportamento que utilizamos o operador `and` na composição do `if`.

# 3.0 - Estrutura de Repetição - `for`

Como vimos na seção anterior, a estrutura `if` nos ajuda a contruir testes lógicos e validações dos valores das nossas variáveis e fazer com que a aplicação siga cursos ou fluxos dependendo de seus valores. Porém, se levarmos em conta o exercício passado, ainda tívemos que fazer muito trabalho manual e repetitivo.

E é aí que entra a estrutura de repetição: Nos possibilitar repeitir uma determinada quantidade de vezes um bloco ou trecho de código. Pegando como exemplo, imagine a base de dados do exercício passado, que tinham somente 8 pacientes. Imagine que essa base possuí agora, 30 pacientes, ou 200 pacientes. Precisaríamos repetir toda essa quantidade de vezes o mesmo bloco para criar a chave `imc` e calcular o seu valor? O quão massante esse processo seria, fora os erros que poderiam ocorrer ao ter que ajustar 200 pacientes? Para resolver esse e outros problemas, utilizamos as estruturas de repetição. 

A estrutura de repetição mais famosa e possivelmente a mais utilizada em Python é o `for`. A estrutura `for` recebe uma estrutura iterável e percorre essa estrutura linha a linha, ou elemento a elemento. Nos permitindo assim manipular elementos indivudualmente dentro de um conjunto, ou, repetir o mesmo bloco de código várias vezes!

Para exemplificar de uma maneira mais simples, vamos pensar em como escrever o nome `Pedro` cinco vezes na tela. Para isso iremos utilizar o `for`:

````python
for indice in range(5):
    print('Pedro')
````

Observe que utilizamos a função `range()` e colocamos o número 5 como argumento. Isso faz com que seja criado um iterador, que irá contar de 0 a 4:

````python
for indice in range(5):
    print(f'O valor da variável índice nesse momento é {indice};')
````

E com isso sabemos que o bloco que está dentro da estrutura `for` será executado 5 vezes. Outro ponto importante é que a variável `indice` sempre irá percorrer elemento a elemento, ou item a item. Esse comportamento pode ser observado no `range()`. Quando passamos o valor `5` como parâmetro, o que a função faz é criar um elemento que tenha os valores de zero até o elemento passado menos um, que no nosso caso gera um elemento que tenham valores de zero a quatro.

Além disso, podemos percorrer, ou iterar, em listas e dicionários também:

````python
# Lista de exemplo
lista = [
    'Pedro',
    32,
    1.83,
    True
]

# Percorrendo a lista utilizando for
for elemento in lista:
    print(f'O elemento da vez é {elemento}')
````

````python
# Dicionário de exemplo
dicionario = { 'nome': 'Pedro', 'idade': 32, 'altura': 1.83, 'casado': True }

# Percorrendo somente as chaves
for chave in dicionario:
    print(f'A chave da vez é {chave}')
````

````python
# Percorrendo somente os valores
for valor in dicionario.values():
    print(f'O Valor da vez é {valor}')
````

````python
# Percorrendo tanto chaves quanto valores
for chave, valor in dicionario.items():
    print(f'A chave da vez é {chave} e o Valor da vez é {valor}')
````

# 4.0 - Funções

Podemos entender funções em Python como blocos que código em que encapsulamos em uma variável! Uma variável não é uma parte da memória que guarda um determinado conteúdo e que possuí um nome? Uma função pode ser entendida da mesma forma: Um pedaço da memória que guarda um conteúdo, que nesse caso é um bloco de código, e que possuí um nome.

Dessa forma, quando criamos uma função, o que estamos fazendo é um bloco de código que pode ser executado várias vezes. As funções tem como objetivo escrevermos código que ou sabemos que vamos repetir várias vezes ao longo da aplicação; ou fará nossa aplicação ficar dividida, fazendo com que cada função seja responsável por uma parte da aplicação, facilitando assim não só a cosntrução e entidimento da aplicação, como também na sua manutenção.

Para exemplificar, podemos pegar o exercício passado. Nele tínhamos várias etapas e várias tarefas:

1. Calcular o IMC para cada paciente da lista.
2. Separar os pacientes que possuem obesidade grau 3 e 2.
3. Entre os pacientes selecionados, calcular a média de altura e peso entre eles.
4. Entre os pacientes selecionados, verificar a distribuição entre casado e não casados
5. Entre os pacientes selecionados, verificar qual é o top 5 de comidas preferidas
6. Entre os pacientes selecionados, criar uma nova chave para cada um dos pacientes. Essa chave deve ter o nome de `age` e o seu conteúdo deve ser a idade do paciente. A idade deve ser calculada com base na data de nascimento dele.

Se pegarmos cada uma dessas tarefas e criarmos uma função para cada uma delas, conseguimos assim separar as responsabilidades de cada tarefa para cada função, tendo assim mais controle em possiveis correções de erros ou melhorias futuras!

Vamos transformar cada uma delas em um função. Para criar um função em Python, utilizamos a seguinte declaração:

```python
def nome_funcao(parametros):
    # Comandos
```

O nome da função pode ser qualquer coisa que desejarmos, mas é importante utilizarmos nomes que nos ajudem a entender o que a função em si está fazendo. Além disso, uma função pode ou não ter parâmetros, tudo vai depender da finalidade e maneira de construção dessa função. Um parametro é uma variável que passamos para a função para que possamos utilizar dentro dela.

Vamos criar as funções para resolvermos as tarefas dos exercícios passados. Primeiro vamos criar a função que calcula o IMC.
````python
# Lendos os dados do arquivo
pacientes = read_csv('csv/rev_p1.csv')
pacientes[0]
````

## 4.1 - Criação da Função de Cálculo do IMC

````python
# Declaração da Função
# ====================
# O nome escolhido foi 'calcula_imc', pois assim
# Fica claro que pra quem for utilizar a função que
# ela irá calcular o IMC de uma lista de pacientes.
# Fora isso, iremos receber a base de dados inteira,
# e por conta disso o parametro foi nomeado como 'pacientes',
# também para indicar que esse parametro conterá vários pacientes
def calcula_imc(pacientes):
    
    # Percorrendo os pacientes recebido no parâmetro
    for paciente in pacientes:

        # Cálculo do IMC
        paciente['imc'] = round(float(paciente['weight_kg']) / (float(paciente['heigth_m']) ** 2), 2)
        
    # Devolve a lista de pacientes com o IMC calculado
    return pacientes
````

````python
pacientes_imc = calcula_imc(pacientes)
pacientes_imc[0]
````

## 4.2 - Separação dos Pacientes com Obesidade grau 2 e 3

````python
def selecao_obesos(pacientes):
    # Criando duas listas que serão utilizadas
    # para receber os pacientes com cada grau de
    # obesidade
    obesidade = []
    
    # Vamos agora percorrer os pacientes e verificar
    # os seus indices de IMC para realizarmos a separação
    for paciente in pacientes:
        
        # Verifica se o paciente possuí IMC maior que 30.
        # Se sim, o paciente possuí obesidade grau 2 ou 3
        if paciente['imc'] > 30:
            obesidade.append(paciente)
            
    # Após percorrer todos os pacientes e selecionar os
    # que possuem obesidade grau 2 e 3, vamos devolver
    # ambas as listas criadas com os pacientes selecionados
    # dentro delas
    return obesidade
````

## 4.3 - Cálculo da Média do Pesos e Altura dos Pacientes

````python
def media_peso_altura(pacientes):
    
    # Essa variável manterá a soma de todos os pesos
    pesos = 0
    
    # Essa variável manterá a soma de todas as alturas
    alturas = 0
    
    for paciente in pacientes:
        
        # Soma as alturas anteriores com a altura do
        # paciente atual
        alturas += float(paciente['heigth_m'])
        
        # Soma os pesos anteriores com o peso do
        # paciente atual
        pesos += float(paciente['weight_kg'])
        
    # Com os pesos todos somados, vamos agora divir a
    # soma das alturas e pesos pela quantidade de pacientes
    return alturas / len(pacientes), pesos / len(pacientes)
````

````python
# Como a função 'media_peso_altura' nos devolve dois
# resultados, temos que recebê-los em variáveis diferentes
media_altura, media_peso = media_peso_altura(pacientes_obesos)
print(f'A media da altura dos pacientes obesos é de {media_altura:.2f} m;')
print(f'A media do peso dos pacientes obesos é de {media_peso:.2f} Kg;')
````

## 4.4 - Distribuição entre os Pacientes casados e não Casados

````python
# Função simples para calcular a porcentagem de um
# de uma determinada quantidade
def c_p(total, quantidade):
    return (quantidade*100) / total
````

````python
def dist_casados(pacientes):
    
    # Dicionário que manterá a contagem e a porcentagem
    # Entre os pacientes obesos
    dist_casados = {
        'casados': {
            'contagem': 0
        },
        'nao_casados': {
            'contagem': 0
        },
    }
    
    # Percorrendo os pacientes
    for paciente in pacientes:
        
        # Verifica se o paciente é casado. Se sim, adiciona
        # em uma unidade a contagem dos pacientes casados
        if paciente['married'].lower() == 'true':
            dist_casados['casados']['contagem'] += 1
            
        # Se não for casado, adiciona em uma unidade a
        # contagem dos pacientes não casados
        else:
            dist_casados['nao_casados']['contagem'] += 1
            
    # Calcula utilizando a função criada anterioremente a porcentagem
    # dos pacientes casados em relação ao total de pacientes
    percent = round(c_p(len(pacientes), dist_casados['casados']['contagem']), 2)
    dist_casados['casados']['porcentagem'] = percent
    
    # Calcula utilizando a função criada anterioremente a porcentagem
    # dos pacientes não casados em relação ao total de pacientes
    percent = round(c_p(len(pacientes), dist_casados['nao_casados']['contagem']), 2)
    dist_casados['nao_casados']['porcentagem'] = percent
    
    # Retorna a estrutura criada com a distribuição entre os pacientes
    # casados e não casados
    return dist_casados
````

````python
distribuicao = dist_casados(pacientes_obesos)
distribuicao
````

## 4.5 - Verifica as Comidas Mais Consumidas

````python
def top_comidas(pacientes):
    
    # Variável responsável por registrar e fazer o
    # ranqueamento das comidas. Iremos utilizar o
    # dicionário para tirar proveito da regra da
    # chave única dessa estrutura: Como só pode haver
    # uma chave no dicionário, cada comida será a chave
    # a sua contagem será o valor
    comidas = {}
    
    # Percorre os pacientes
    for paciente in pacientes:
        
        # Recupera o nome da comida que o paciente
        # mais come e salva ela em uma variável
        comida = paciente['food']
        
        # Veririfica se a comida que o paciente mais
        # como já está inserida como chave no dicionário
        # de comidas. Se estiver, soma mais uma unidade
        # ao valor da chave
        if comida in comidas.keys():
            
            # Soma mais uma unidade à comida
            comidas[comida] += 1
            
        # Se a comida ainda não estiver dentro do dicionário
        # Vamos inserí-la com o valor zero
        else:
            comidas[comida] = 0
            
    # Com todas os pacientes percorridos e as comidas já
    # contabilizadas, vamos organizar esse dicionário para
    # que a comida com a maior contagem decrescente
    return dict(sorted(comidas.items(), key=lambda comida: comida[1], reverse=True))
````

````python
comidas = top_comidas(pacientes_obesos)
print(list(comidas.keys())[:5])
````

## 4.6 - Calcula a Idade de Cada Paciente

Por fim, vamos criar uma nova função que irá criar a chave `age` nos nossos pacientes. Para isso teremos que utilizar o módulo `datetime` do pacote `datetime`:

````python
from datetime import datetime


def calc_age(pacientes):

    # Percorrendo a os pacientes
    for paciente in pacientes:
        
        # Recuperando a data atual utilizando o pacote
        # datetime
        today = datetime.today()

        # Convertendo a data de nascimento do paciente 
        # em um objeto datetime para que possamos fazer
        # o cálculo da data
        birth = datetime.strptime(paciente['birth'], '%d-%m-%y')

        # Faz o o cálculo da idade e armazena na chave
        # 'age' do paciente atual
        paciente['age'] = today.year - birth.year - ((today.month, today.day) < (birth.month, birth.day))

    # Retorna a lista de pacientes com a idade calculada
    return pacientes
````

````python
pacientes_obesos = calc_age(pacientes_obesos)
pacientes_obesos[0]
````

# 5.0 - Programação Orientada à Objetos - POO

Agora que ja revimos o básico do Python, vamos para um assunto um pouco mais complexo, mas igualmente simpels. Orientação à Objeto. Programação Orientada à Objetos é um paradigma de desenvolvimento de sistemas quem possuí várias premissas, sendo algumas delas economia de código e facilidade de manutenção e escalonamento de aplicação, dentre outras. É um dos paradigmas mais utilizados no dia-a-dia de desenvolvimento de sistemas, e precisamos aprender o básico dele pois, além do Python utilizar esse paradigma em sua construção, muitas bibliotecas de ciencia de dados utilizam ele.

A idéia central da Programação Orientada à Objetos, ou POO, é fazer com que criemos representações de entidades do mundo real, como carros, cadeiras, animais, modelos de machine learning, tenham uma "contra-parte" computacional. Ou seja, descrevemos as características e funcionalidades de um determinado objeto do mundo real, para o mundo computacional.

Quando descrevemos essas características, o que estamos fazendo é criar o que chamamos de **classe**. Uma classe é um molde ou como um modelo/exemplo, que, através desse molde ou exemplo, criamos nossos objetos. É na classe que descrevemos quais características, ou atributos, e quais funcionalidades, ou métodos, a nossa representação terá. Além disso, é através da nossa classe que criamos os nossos objetos.

Para fazer uma outra analogia, imagine que temos uma massa de biscoitos aberta em nossa mesa e queremos cortar a massa e formato de estrela. O molde de metal que utilizamos para cortar os biscoitos em formato de estrela é a classe. Já os objetos são os biscoitos em formato de estrela que cortamos da nossa massa.

Por tanto, a classe é uma representação de um entidade do mundo real, que contém atributos e métodos, e é através da classe que criamos os nossos objetos. 

Para exemplificar como criar classes e objetos em Python, vamos pensar em uma entidade do mundo real: Um Cientista de Dados. Vamos imaginar que todo estamos ajudando a contruir o software de cadastro de colaboradores da nossa empresa. Nele vamo criar a classe `CientistaDados`, contendo os seus atributos e métodos. Todas as informações são os nossos atributos, ou seja, as características do nosso cientista de dados.

Fora isso, os cientistas de dados da empresa podem codificar (escrever código), criar modelos de Machine Learning e receber promoções pelo bom trabalho apresentado. # 6.0 - Tratamento de Erros

Para finalizar a nossa revisão, vamos verificar como os erros acontecem no Python e como podemos tratá-los. Primeiro, temos que entender que o Python é uma linguagem de programação Hibrida, ou seja, ela é compilada e interpretada ao mesmo tempo. O que faz com que basicamente existam dois tipos de erros básicos:
* Erros de Compilação
* Erros Lógicos

**Erros de Compilação** são os erros que ocorrem quando escrevemos palavras erradas, tentamos somar uma String com um int, erramos a indentação, tentamos recuperar o dado de um indice que não existe de uma lista/tupla e etc. No geral, são erros que comentemos e que o compilador do Python detecta ao verificar o que escrevemos.

**Erro de Lógica** são os erros que ocorrem quando o compilador do Python não encontra nada de errado em nossa escrita/código e executa o que está escrito, porém, o resultado é algo inesperado por nós, seja um valor calculado errado ou um loop infinito que criamos.

Por tanto, todo erro que cometemos, de uma forma básica, pode ser enquadrado em uma dessas duas categorias. Um "guia" que pode ser utilizado para nos ajudar a verificarmos e a entendermos melhor os nossos erros, pode ser observado nas figuras abaixo:

## 6.1 - Função `assert`

A função `assert` no Python tem como objetivo criar um ponto onde seja verificado um determinado teste e se esse teste for verdadeiro, a aplicação segue o fluxo normal de execução. Porém, se o teste não for verdadeiro, a aplicação é encerrada. Pense no assert como um `if` que ou faz com que a aplicação siga o seu fluxo, ou faz com que a aplicação seja encerrada.

Podemos testar várias coisas utilizando o `assert`, como se um adeterminada variável é de um determinado tipo, ou se o resultado de uma determinada função é o esperado, etc. Porém, ela é mais utilizada em classes e funções de teste, que geralmente são classes e funções que estão fora da aplicação principal e são criadas exatamente para testarmos como uma parte da aplicação está funcionando. O famoso **teste unitário**.

Dessa forma, não é recomendado que se utilize a função `assert` dentro do bloco ou classes principais da aplicação, justamente pela capacidade que ela possuí de parar a aplicação caso haja algum erro.

Para utilizar a função `assert`, basta seguir o seguinte modelo:

```python
assert teste, 'Mensagem de erro'
```

Onde o `teste` pode ser o resultado de uma função, conteúdo de uma variável e etc. Por exemplo, vamos criar uma função simples que retorne a soma entre dois valores: 

````python
def soma(val1, val2):
    return val1 + val2
````

Agora vamos testar o resultado dessa função. Digamos que eu tenha criado duas variáveis, `a` e `b`, que recebem dados digitados pelo usuário, e que a variável `res` receberá a soma desses dois valores:

````python
a = int(input('Informe o primeiro valor: '))
b = int(input('Informe o segundo valor: '))
````

````python
res = soma(a, b)
res
````

Para utilizarmos a função `assert` para testarmos a resposta, podemos fazer da seguinte forma:

````python
assert soma(a, b) == a + b, 'Função está somando errado'
````

Ou ainda:

````python
assert soma(5, 8) == 13, 'Função está somando errado'
````

Como não estamos recebendo nenhum tipo de erro ou mensagem, o teste funcionou corretamente. Caso a função estivesse com um comportamento errado, ou inesperado, receberíamos a mensagem que está após a virgula, e a aplicação seria encerrada:

````python
assert soma(5, 8) == 1, 'Função está somando errado'
````

Todo erro produzido pela função `assert` sempre gerará um erro de `AssertioError`. E se não colocarmos nenhuma mensagem de erro, a função `assert` simplismente devolverá o `AssertionError`:

````python
assert soma(5, 8) == 1
````

## 6.2 - Função `try .. except`

Outra técnica que podemos utilizar é a função `try .. except`. Essa função nos permite tentar executar o bloco de código que está dentro do `try` , caso haja alguma falha na execução desse código que faça a aplicação pare, será executado o bloco `except`.

No geral, utilizamos essa estrutura quando estamos trabalhando com instruções sensíveis, como leitura ou gravação de arquivos, operações que possam gerar uma divisão com zero, muitas transformações em variáveis e etc.

Ao contrário da função `assert`, a função `try .. except` não finaliza a aplicação, o que faz com que seja possível controlarmos os possíveis erros em tempo de execução da nossa aplicação e contorná-los.

Para exemplificar, digamos que temos as seguintes listas de valores:

````python
lista_valores_1 = [1, 3, 5, 0, 0, 3, 4]
lista_valores_2 = [5, 6, 1, 8, 0, 1, 2]
````

E que precisamos a realizar a divisão desses dos valores que estão na primeira lista (`lista_valores_1`) pelos valores que estão na segunda lista (`lista_valores_2`):

````python
print(f'Mensagem de fluxo! Antes da execução do for\n')

for indice in range(len(lista_valores_1)):
    print(f'O resultado da divisão é {lista_valores_1[indice] / lista_valores_2[indice]}')
    
print(f'\nMensagem de fluxo! Depois da execução do for')
````

Observe que a mensagem `print(f'\nMensagem de fluxo! Depois da execução do for')` não foi executada pois houve um erro na operação.

Imagine que os valores de ambas as listas sejam todos aleatórios e sejam provenientes do uso dos usuários do sistema. Se fizermos a divisão como está na célula acima, e ocorrer de haver um zero na segunda lista, haverá uim erro na aplicação que fará com que ela pare de funcionar, quebrando assim o sistema/aplicação.

Para evitar esse erro, podemos utilizar a função `try .. except` para tentar fazer a operação. Caso a aplicação tente fazer uma divisão com zero, ela não irá para, e sim executará o bloco que está dentro do `except`:

````python
# Lista de valores
lista_valores_1 = [1, 3, 5, 0, 0, 3, 4]
lista_valores_2 = [5, 6, 1, 8, 0, 1, 2]

print(f'Mensagem de fluxo! Antes da execução do try .. except\n')

try:
    print(f'Mensagem de fluxo! Antes da execução do for\n')

    for indice in range(len(lista_valores_1)):
        print(f'O resultado da divisão é {lista_valores_1[indice] / lista_valores_2[indice]}')

    print(f'\nMensagem de fluxo! Depois da execução do for')
except:
    print(f'\nMensagem de ERRO! Houve algum erro na execução do Código')
    print(f'O bloco except foi executado ao invés do try')
    
    
print(f'\nMensagem de fluxo! Depois da execução do try .. except')
````

Observe que agora a aplicação em si não parou sua execução quando o erro foi encotrado. Ao invés disso, a aplicação continuou e executou o que está descrito no bloco `except` e continou o seu fluxo de execução!

Dessa forma, qualquer operação sensível e que tenha a possibilidade de gerar erros estiver sendo escrita, podemos utilizar o bloco `try .. except` para trabalhar com possívies erros e fazer com que a aplicação pare ou seja interrompida.

Além disso, podemos veriricar quais erros ocorreram e criar cláusulas `except` para diferentes tipos de exceções. Por exemplo, no caso anterior, poderíamos escrever os erros:

````python
# Lista de valores
lista_valores_1 = [1, 3, 5, 0, 0, 3, 4]
lista_valores_2 = [5, 6, 1, 8, 0, 1, 2]

# Import da biblioteca para verificação das exceções
import sys

print(f'Mensagem de fluxo! Antes da execução do try .. except\n')

try:
    print(f'Mensagem de fluxo! Antes da execução do for\n')

    for indice in range(len(lista_valores_1)):
        print(f'O resultado da divisão é {lista_valores_1[indice] / lista_valores_2[indice]}')

    print(f'\nMensagem de fluxo! Depois da execução do for')
except:
    print(f'\nMensagem de ERRO! Houve algum erro na execução do Código')
    print(f'Erro: {sys.exc_info()[0]}. Motivo: {sys.exc_info()[1]}')
    print(f'O bloco except foi executado ao invés do try')
    
    
print(f'\nMensagem de fluxo! Depois da execução do try .. except')
````

E como dito anteriormente, podemos também gerar cláusulas exept para erros distintos. Por exemplo:

````python
# Lista de valores
lista_valores_1 = [1, 3, '5', 0, 0, 'P', 4]
lista_valores_2 = [5, 6, 1, 8, 0, 1, 2]

# Import da biblioteca para verificação das exceções
import sys

print(f'Mensagem de fluxo! Antes da execução do try .. except\n')

try:
    print(f'Mensagem de fluxo! Antes da execução do for\n')

    for indice in range(len(lista_valores_1)):
        print(f'O resultado da divisão é {lista_valores_1[indice] / lista_valores_2[indice]}')

    print(f'\nMensagem de fluxo! Depois da execução do for')
except(ZeroDivisionError):
    print(f'\nMensagem de ERRO! Tentaiva de divisão por zero')
except:
    print(f'\nMensagem de ERRO! Houve algum erro na execução do Código')
    print(f'Erro: {sys.exc_info()[0]}. Motivo: {sys.exc_info()[1]}')
    print(f'O bloco except foi executado ao invés do try')
    
    
print(f'\nMensagem de fluxo! Depois da execução do try .. except')
````

Observe que temos duas cláusulas do bloco `except`: Uma para divisão por zero e outra para qualquer outra exceção que não seja divisão por zero. Observe também que o `except` executado foi o geral, onde o tipo de exeção não foi definido.

````python
class Aluno:
    
    def __init__(self, nome, nota1 = 0, nota2 = 0, nota3 = 0, nota4 = 0):
        self.nome = nome
        self.nota1 = nota1
        self.nota2 = nota2
        self.nota3 = nota3
        self.nota4 = nota4
        
        
    def calcula_media(self):
        return (self.nota1 + self.nota2 + self.nota3 + self.nota4) / 4
    
    def aprovado(self):
        if self.calcula_media() >= 7:
            return 'Aprovado'
        else:
            return 'Reprovado'
        
    def __str__(self):
        return f'O aluno {self.nome} tem as notas = {self.nota1}, {self.nota2}, {self.nota3}, {self.nota4}. A media {self.calcula_media()} e está {self.aprovado()}')
````
